```python
import re
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import pandas as pd

# Daftar kata tidak relevan (stopwords tambahan)
meaningless_words = {'salah', 'proses', 'penelitian', 'hasil', 'memiliki', 'dengan', 'untuk', 'dari', 'adalah', 'yang', 'akan',
                     'dalam', 'sebagai', 'atau', 'ke', 'di', 'pada', 'oleh', 'juga', 'ini', 'itu', 'tersebut', 'pengguna', 'penulis',
                     'pemain', 'kucing', 'kegunaan', 'inventaris','inventory', 'pakar','kulit','kerena', 'oleh','sma'}

# Membaca data normalisasi
normalisasi_df = pd.read_csv('/content/drive/MyDrive/Bahan Skripsi/PERSIAPAN PENDADARAN/DATAFINAL/normalisasi.csv')

# Menambahkan istilah baru untuk normalisasi
new_term_df = pd.DataFrame({'Istilah Umum': ['prose','medium'], 'Istilah Normalisasi': ['proses','media']})
normalisasi_df = pd.concat([normalisasi_df, new_term_df], ignore_index=True)

# Membuat kamus untuk normalisasi
normalisasi_dict = dict(zip(normalisasi_df['Istilah Umum'].str.lower(), normalisasi_df['Istilah Normalisasi'].str.lower()))

# Inisialisasi lemmatizer
lemmatizer = WordNetLemmatizer()

# Fungsi untuk normalisasi teks
def normalize_text(text):
    return ' '.join([normalisasi_dict.get(word, word) for word in text.split()])

# Fungsi untuk preprocessing teks
def preprocess_text(text):
    # Hapus URL
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    # Hapus angka
    text = re.sub(r'\d+', '', text)
    # Hapus tanda baca
    text = re.sub(r'[^\w\s]', '', text)
    # Ubah menjadi huruf kecil
    text = text.lower()
    # Normalisasi teks
    text = normalize_text(text)
    # Tokenisasi teks
    tokenizer = RegexpTokenizer(r'\w+')
    words = tokenizer.tokenize(text)
    # Menggunakan hanya stopwords Bahasa Indonesia dan kata-kata tidak relevan
    stop_words = set(stopwords.words('indonesian')).union(meaningless_words)
    filtered_words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words and len(word) > 2]
    return ' '.join(filtered_words)

# Menerapkan preprocessing ke kolom 'abstract' dan menyimpannya ke kolom baru
skripsi['cleaned_abstract'] = skripsi['abstract'].apply(preprocess_text)
```

##      MODELING LDA
========================

Membuat dictionary dan corpus

```python
tokenized_abstracts = skripsi['cleaned_abstract'].apply(lambda x: x.split())
id2word = corpora.Dictionary(tokenized_abstracts)
corpus = [id2word.doc2bow(text) for text in tokenized_abstracts]
```

```python
alpha_list = ['auto', 'symmetric', 'asymmetric']
beta_list = [0.01, 0.1, 0.5, 0.6]
results = []

# Iterasi untuk menemukan kombinasi terbaik dari alpha, beta (eta), dan num_topics
for num_topics in range(2, 21):
    for alpha in alpha_list:
        for beta in beta_list:
            print(f"Training LDA dengan num_topics={num_topics}, alpha={alpha}, dan beta={beta}...")
            lda_model = gensim.models.LdaModel(corpus=corpus,
                                                id2word=id2word,
                                                num_topics=num_topics,
                                                random_state=42,
                                                update_every=1,
                                                chunksize=300,
                                                passes=20,
                                                alpha=alpha,
                                                eta=beta,
                                                per_word_topics=True)
            coherence_model_lda = CoherenceModel(model=lda_model,
                                                 texts=tokenized_abstracts,
                                                 dictionary=id2word,
                                                 coherence='c_v')

            coherence_score = coherence_model_lda.get_coherence()
            print(f'Koherensi: {coherence_score:.4f} untuk num_topics={num_topics}, alpha={alpha}, dan beta={beta}')
            results.append((num_topics, alpha, beta, coherence_score))
results_df = pd.DataFrame(results, columns=['Num Topics', 'Alpha', 'Beta', 'Coherence'])
best_result = results_df.loc[results_df['Coherence'].idxmax()]
print(f"\nHasil terbaik:")
print(f"Koherensi terbaik: {best_result['Coherence']:.4f}")
print(f"Num Topics terbaik: {best_result['Num Topics']}")
print(f"Alpha terbaik: {best_result['Alpha']}")
print(f"Beta (eta) terbaik: {best_result['Beta']}")
```

```python
num_topics_best = 5
alpha_best = 'auto'
beta_best = 0.6
lda_model_best = gensim.models.LdaModel(
    corpus=corpus,
    id2word=id2word,
    num_topics=num_topics_best,
    random_state=42,
    update_every=1,
    chunksize=300,
    passes=20,
    alpha=alpha_best,
    eta=beta_best,
    per_word_topics=True
)
topics = lda_model_best.print_topics(num_words=6)
print("\nTopik yang ditemukan oleh model LDA dengan parameter terbaik:")
for topic in topics:
    print(topic)
```

```python
coherence_model_lda = CoherenceModel(model=lda_model_best, texts=tokenized_abstracts, dictionary=id2word, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print(f"\nNilai Koherensi LDA untuk {num_topics_best} topik: {coherence_lda:.4f}")
```

```python
def bert_topic_modeling(preprocessed_texts):
    coherence_values = []
    min_topic_sizes = range(2, 21)

    for min_topic_size in min_topic_sizes:
        model = BERTopic(min_topic_size=min_topic_size, verbose=False, embedding_model='all-MiniLM-L6-v2')
        topics, _ = model.fit_transform(preprocessed_texts)
        topic_words = model.get_topics()
        topic_list = []
        for topic_id, words in topic_words.items():
            if topic_id != -1:
                topic_list.append([word[0] for word in words])
        coherence_model = CoherenceModel(topics=topic_list, texts=tokenized_abstracts.tolist(), dictionary=id2word, coherence='c_v')
        coherence = coherence_model.get_coherence()
        coherence_values.append(coherence)
        print(f"min_topic_size: {min_topic_size}, Koherensi: {coherence:.4f}")
    best_index = np.argmax(coherence_values)
    best_min_topic_size = min_topic_sizes[best_index]
    print(f"\nJumlah topik terbaik (min_topic_size): {best_min_topic_size} dengan koherensi: {coherence_values[best_index]:.4f}")
    return best_min_topic_size
best_min_topic_size = bert_topic_modeling(skripsi['cleaned_abstract'].tolist())
```

```python
# Building the BERTopic Model with Best min_topic_size
best_model = BERTopic(min_topic_size=best_min_topic_size, verbose=True, embedding_model='all-MiniLM-L6-v2')
headline_topics, _ = best_model.fit_transform(skripsi['cleaned_abstract'].tolist())
```
